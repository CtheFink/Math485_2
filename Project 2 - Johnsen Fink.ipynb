```python
dfs = [df2005, df2008, df2011, df2014, df2017, df2021, df2023] 

data = np.array([df.values for df in dfs])

print("3D Array Shape:", data.shape)
```

```python
def remove_commas(data):
    # Use vectorized operation to replace commas, ensuring we only process strings
    cleaned_data = np.vectorize(lambda x: x.replace(',', '').strip() if isinstance(x, str) else x)(data)
    return cleaned_data
data=remove_commas(data)
print(data)
```

```python
def safe_convert(data):
    return np.array(data, dtype=int)

data = safe_convert(data)

for i in range(data.shape[0]):
    np.fill_diagonal(data[i], 0)

fig, ax = plt.subplots()
heatmap = ax.imshow(data[0], cmap='plasma', interpolation='nearest')

def update(frame):
    heatmap.set_array(data[frame])
    return [heatmap]
ani = animation.FuncAnimation(fig, update, frames=data.shape[0], blit=True)

ani.save('moving_heatmap.gif', writer='pillow')

plt.show()
```
