```python
df.rename(columns = {'mci_uniq_id': 'id', 
                    'date_of_event': 'date', 
                    'marital_status': 'marital', 
                    'education_level':'education'}, inplace=True)
def get_recipient_attribute(df):
    recipient = df.groupby(['id']).agg(
        num_service = ('service', 'count'),
        distinct_service = ('service', 'nunique'), 
        first_date = ('date', 'min'), 
        last_date = ('date', 'max'), 
        num_month = ('month', 'count'), 
        distinct_month = ('month', 'nunique')
    ).reset_index()

    return recipient
```

```python
#Step 2,3,4 In-Class Exercise

def get_service_attributes(df):
    service_attributes = df.groupby('service').agg(
        unique_id=('MCI_UNIQ_ID', 'nunique'),
        total_services=('service', 'count'),
        diff_services=('service', 'nunique'),
        ).reset_index()
    return service_attributes

service_attributes_df = get_service_attributes(df)

print(service_attributes_df)
#It appears that reset_index() only collapses the data to make it more condense. Without it, it has to reprint all of the services/list all the services.
```

```python
def get_service_attributes(df):
    sorted_data = df.groupby(['service', 'DATE_OF_EVENT'])['MCI_UNIQ_ID'].nunique().reset_index()
    service_attributes = sorted_data.groupby('service').agg({
        'MCI_UNIQ_ID': ['sum', 'mean']
    })
    service_attributes.columns = ['total_recipients', 'avg_monthly_recipients']
    
    return service_attributes

service_attributes_df = get_service_attributes(df)

print(service_attributes_df)
#Returns a list of services, the total numebr of recipients per service, and the average monthly number of recipients per service.
```
